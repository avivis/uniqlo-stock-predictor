This is a machine learning-based stock predictor created using pandas and Facebook's Prophet for time series forecasting on Uniqlo stock closing price data.

```python
#importing pandas
import pandas as pd
```

```python
#importing uniqlo stock price data
from google.colab import files
files.upload()
```

```python
df = pd.read_csv('uniqlostockdata.csv')
df
```

```python
#getting date and close price columns from data
df = df[['Date', 'Close']]
df = df.rename(columns = {'Date': 'ds', 'Close': 'y'})
df
```

```python
#extracting testing set from data
last = df.head(20)
last
```

```python
#extracting training set from data
df = df.iloc[20: , :]
df
```

```python
#getting facebook prophet library and creating and training model
from fbprophet import Prophet
fbp = Prophet(daily_seasonality = True)
fbp.fit(df)
future =fbp.make_future_dataframe(periods=365)
forecast = fbp.predict(future)
```

```python
#plotting prediction model
from fbprophet.plot import plot_plotly, plot_components_plotly
plot_plotly(fbp, forecast)
```

```python
#see models prediction on an index from testing set
forecast[forecast.ds == '2016-12-22']['yhat']
```
